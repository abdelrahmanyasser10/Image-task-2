```python
import cv2 
import matplotlib.pyplot as plt
import numpy as np
```

```python
def PixelColorExtraction(event,x,y,flag,param):
    global pixel,hsv_Deadpool,Deadpool
    if event == cv2.EVENT_LBUTTONDOWN:
        pixel = (y, x)
        pixel_color = hsv_Deadpool[pixel]
        lower_bound = np.maximum(pixel_color, 0)
        upper_bound = np.minimum(pixel_color, 255)
        extracted_image = cv2.inRange(hsv_Deadpool, lower_bound, upper_bound)
        masked_image = cv2.bitwise_and(Deadpool, Deadpool, mask=extracted_image)
        display_img(extracted_image)
        display_img(masked_image)
```

```python
def display_img(img):
    new_img= cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(new_img)
    plt.show()
```

```python
pixel = None
Deadpool = cv2.imread('unnamed.jpg')
hsv_Deadpool = np.array(cv2.cvtColor(Deadpool, cv2.COLOR_BGR2HSV))


while(1):
    cv2.namedWindow('Deadpool')
    cv2.imshow('Deadpool',Deadpool)
    cv2.setMouseCallback('Deadpool',PixelColorExtraction)
    key = cv2.waitKey(1)
    if key == 27:
        break
cv2.destroyAllWindows()
```
